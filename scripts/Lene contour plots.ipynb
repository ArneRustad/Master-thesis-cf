```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def create_contour_for_loss(loss_func, xlabel='feature_x', ylabel='feature_y', n_grid=50):
    feature_x = np.linspace(0, 6, n_grid)
    feature_y = np.linspace(0, 6, n_grid)
  
    # Creating 2-D grid of features
    [X, Y] = np.meshgrid(feature_x, feature_y)
    fig, ax = plt.subplots(1, 1)

    Z = loss_func(X, Y)

    # plots contour lines
    contour_mapping = ax.contour(X, Y, Z, cmap="viridis", levels=np.linspace(np.min(Z), np.max(Z), 20))

    ax.set_title('Contour Plot')
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    plt.colorbar(contour_mapping, ax=ax)

    plt.show()
```

```python
loss_func1 = lambda x,y: abs(x + y - 5)
create_contour_for_loss(loss_func1)
```

```python
loss_func2 = lambda x,y: (x + y - 5)**2
create_contour_for_loss(loss_func2)
```

```python
loss_func3 = lambda x,y: abs(x + y - 5) + 0.1*abs(x - 2.5) + 0.1*abs(y - 2.5)
create_contour_for_loss(loss_func3)
```

```python
n = 1000
corr = 0.8
beta1 = 2
beta2 = 2
drawn_bivariate_samples = np.random.multivariate_normal(mean=[0, 0], cov=[[1, corr], [corr, 1]], size=n)
z1, z2 = drawn_bivariate_samples[:, 0], drawn_bivariate_samples[:, 1]
true_mean = z1 * beta1 + z2 *beta2
response = true_mean + np.random.normal(loc=0, scale=2,size=n)
```

```python
loss_func_regression = np.vectorize(lambda beta1, beta2: np.mean((response - (z1 * beta1 + z2 * beta2))**2))
create_contour_for_loss(loss_func_regression, xlabel=r"$\beta_1$", ylabel=r"$\beta_2$")
```
