```python
import pandas as pd
import numpy
covtype = pd.read_csv(os.path.join(const.dir.data(), "comparison", "covtype.csv"))
covtype_edited = covtype.iloc[:, ~ covtype.columns.str.contains("^(Soil_Type|Wilderness_Area)", regex=True)]
covtype_edited.insert(0, "Soil_Type", covtype.loc[:, covtype.columns.str.startswith("Soil_Type")].idxmax(axis=1))
covtype_edited.insert(0, "Wilderness_Area", covtype.loc[:, covtype.columns.str.startswith("Wilderness_Area")].idxmax(axis=1))
covtype_edited.loc[:, "Cover_Type"] = "CoverType" + covtype.loc[:, "Cover_Type"].astype("string")
covtype_edited.to_csv(os.path.join(const.dir.data(), "comparison", "covtype_edited.csv"), index=False)
covtype_edited
```

```python
import pandas as pd
import numpy
news = pd.read_csv(os.path.join(const.dir.data(), "comparison", "OnlineNewsPopularity.csv"))
news.columns = [column.strip() for column in news.columns]
news_edited = news.loc[:, ~ news.columns.str.contains("^(weekday_is|data_channel)")]
news_edited = news_edited.drop("url", axis=1)
news_edited.insert(
    0, "weekday",
    news.loc[:, news.columns.str.contains("^weekday_is")].idxmax(axis=1).str.replace("weekday_is_", "").str.title()
)
news_edited.insert(0, "data_channel",
                   news.loc[:, news.columns.str.contains("^data_channel")].idxmax(axis=1).str.replace("data_channel_is_", ""))
#news_edited.loc[:, "Class"] = "Class" + news.loc[:, "Class"].astype("string")
news_edited.loc[:, "is_weekend"] = np.where(news_edited.loc[:, "is_weekend"] > 0.5, "yes", "no")
news_edited.to_csv(os.path.join(const.dir.data(), "comparison", "news_edited.csv"), index=False)
news_edited
```

```python
import sklearn.model_selection
import helpers
covtype_train, covtype_test = sklearn.model_selection.train_test_split(covtype_edited, train_size=0.7)
metrics, cats = helpers.eval.fit_and_evaluate_xgboost(covtype_train, covtype_test, response_col="Cover_Type",
                                                    retcats=True, metrics=["accuracy", "f1"])
metrics
```

```python
import sklearn.model_selection
import helpers

def transform_news_dataset(news):
    return news.assign(shares = lambda x: np.where(x.shares > 3395, "Popular", "Unpopular"))

news_train, news_test = sklearn.model_selection.train_test_split(news_edited, train_size=0.7)
metrics, cats = helpers.eval.fit_and_evaluate_xgboost(news_train, news_test, response_col="shares",
                                                    retcats=True, metrics=["accuracy","auc", "f1", "f1_0", "f1_1"],
                                                     data_transform=transform_news_dataset)
metrics
```

```python
from tabGAN import TabGAN
tg_qtr = TabGAN(news_edited, batch_size=500, ctgan=True, n_critic=10, adam_beta1 = 0.5,
                 quantile_transformation_int=True, quantile_rand_transformation=True, qtr_spread=1,
                 activation_function="mish", jit_compile=True, qt_n_subsample=int(1e5))
```

```python
tg_qtr.train(300, progress_bar=True)
```

```python
metrics= helpers.eval.fit_and_evaluate_xgboost(tg_qtr.sample(), news_test, response_col="shares",
                                                      metrics=["accuracy","auc", "f1", "f1_0", "f1_1"],
                                                     data_transform=transform_news_dataset)
metrics
```
